```python
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Jhonnas19saldua/1st-avtivity/main/Applicant-details.csv')
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df['Residence_City']= df['Residence_City'].astype('category')
```

```python
df['Residence_City'].info()
```

```python
df['Residence_City'].head()
```

```python
df['Annual_Income']=pd.to_numeric(df['Annual_Income'],errors='coerce').astype('Int64')
```

```python
df['Annual_Income'].info()
```

```python
df.head()
```

```python
df['Work_Experience']=pd.to_datetime(df['Work_Experience'],errors='coerce')
```

```python
df['Work_Experience'].info()
```

```python
df['Years_in_Current_Employment'].info()
```

```python
df['Applicant_Age'] = df['Applicant_ID'] - df['Annual_Income']
```
